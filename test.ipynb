```python
from PyPDF2 import PdfMerger

def combine_pdfs(pdf_list, output_filename):
    # Create a PdfMerger object
    merger = PdfMerger()
    
    # Append each PDF file from the list
    for pdf in pdf_list:
        merger.append(pdf)
    
    # Write out the merged PDF to the output file
    merger.write(output_filename)
    merger.close()

if __name__ == '__main__':
    # List your PDF files here; ensure the files are in the same directory as the script,
    # or provide full paths to the files.
    pdf_files = ['file1.pdf', 'file2.pdf']
    
    # Name of the output file
    output_file = 'combined.pdf'
    
    # Combine the PDFs
    combine_pdfs(pdf_files, output_file)
    
    print(f"PDFs have been successfully combined into {output_file}")
```
